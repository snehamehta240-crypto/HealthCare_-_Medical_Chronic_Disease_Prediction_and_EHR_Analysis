```python
# BIOMAKER DISTTRIBUTION PLOT
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
dt=pd.read_csv("HCP.csv")

sns.histplot(df["age"])
plt.title("age distribution")
plt.savefig("age_distribution.png")
plt.show()
```

```python
print(dt.duplicated().sum())
```

```python
print(dt["classification"].unique())
```

```python
# Remove hidden spaces/tabs/newlines
dt['classification'] = dt['classification'].str.strip()

# Lowercase for uniformity
dt['classification'] = dt['classification'].str.lower()
```

```python
# Blood Pressure Distribution
sns.histplot(dt["bp"],kde=True)
plt.title("Blood Pressure Distribution")
plt.savefig("blood_pressure_distribution.png")
plt.show()
```

```python
sns.violinplot(x="classification",y="bp",data = dt,color="red")
plt.title("Blood Pressure Distribution: CKD vs Non-CKD")
plt.show()
```

```python
# sugar Distribution 
# Overall distribution
sns.histplot(dt["su"], kde=True)
plt.title("Sugar_Distribution")
plt.savefig("sugar_distribution.png")
plt.show()

```

```python
# CKD vs Non-CKD comparison
sns.violinplot(x="classification", y="su", data=dt)
plt.title("Sugar Levels: CKD vs Non-CKD")
plt.show()
```

```python
# SERUM CREATININE
# SERUM CREATININE DISTRIBUTION
sns.histplot(dt["su"], kde=True)
plt.title("Serum Creatinine Distribution")
```

```python
#Serum Creatinine: CKD vs Non-CKD
sns.violinplot(x="classification", y="sc", data=dt,color="blue")
plt.title("Serum Creatinine: CKD vs Non-CKD")
plt.show()
```

```python
# OVERALL DISTRIBUTION OF HEMOGLOBIN
sns.histplot(dt["hemo"], kde=True)
plt.title("Hemoglobin Distribution")
plt.savefig("hemoglobin_distribution.png")
plt.show()
```

```python
# HEMOGLOBIN LEVELS DISTRIBUTION : CKD VS NON-CKD
sns.violinplot(x="classification", y="hemo", data=dt)
plt.title("Hemoglobin Levels: CKD vs Non-CKD")
plt.show()
```

```python
#  HYPERTENSION DISTRIBUTION 
sns.countplot(x="htn", hue="classification", data=dt)
plt.title("Hypertension Frequency: CKD vs Non-CKD")
plt.show()
```

```python
# DIABETES DISTRIBUTION :-
sns.barplot(x="dm", hue="classification", data = dt )
plt.title("DIABETES DISTRIBUTION : CKD VS NON-CKD")
plt.show( )
```

```python
sns.histplot(dt["dm"])
plt.show()
```

```python
# BIOMARKER COMPARE
plt.scatter(dt["age"], dt["bp"])
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.title("Age vs Blood Pressure")
plt.savefig("age_vs_bp.png")
plt.show()
```

```python
# age vs serum creatine 
plt.scatter(dt["age"],dt["sc"])
plt.xlabel("age")
plt.ylabel("serum creatinine")
plt.title("age vs serum creatinine ")
plt.savefig("age_vs_serum_creatinine.png")
plt.show()
```

```python
# CORRELATION MATRIX
corr_matrix = dt.corr(numeric_only=True)
print(corr_matrix)
corr_matrix.to_csv("correlation_matrix.csv")
```

```python
# CORRELATION HEATMAP
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix,annot=True,cmap="coolwarm")
plt.title("CORRELATION HEATMAP")

plt.savefig("CORRELATION_HEATMAP.png",dpi=300,bbox_inches="tight")
plt.show()
```
